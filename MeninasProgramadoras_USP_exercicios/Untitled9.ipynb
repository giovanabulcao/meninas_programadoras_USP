```python
caractere = input()

especial = '@#$%&*()_-+=!'
vogais = 'AEIOUaeiou'
algarismo = '0123456789'

if caractere in especial:
  print('especial')
elif caractere in vogais:
  print('vogal')
elif caractere in algarismo:
  print('algarismo')
else:
  print('outro')
```

```python
id1 = int(input())
id2 = int(input())
id3 = int(input())
id4 = int(input())

mvelha = id1

if mvelha < id2: #id2 maior que id1(mvelha), entao mvelha = id2..
   mvelha = id2
if mvelha < id3:
   mvelha = id3
if mvelha < id4:
   mvelha = id4

print(f'A mais velha tem {mvelha} anos')
```
