```python
chute = int(input('digite um valor de chute: ' ))
for v_aleatorio in range(1,11):
    if v_aleatorio < chute:
        result = 'Chutou alto'
    elif v_aleatorio > chute:
        result = 'Chutou baixo'
    else:
        result = 'Acertou!'
        print(result)
print(result,'tente novamente.')
```
