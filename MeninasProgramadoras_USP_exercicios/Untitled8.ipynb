```python
hrs = int(input())
min = int(input())

total = hrs*60 + min
ciclos = total // 90

if ciclos >= 5 or ciclos >= 6:
  print(f'{ciclos} ciclos indica uma boa noite de sono')
else:
  print(f'{ciclos} não indica uma boa noite de sono')
```
