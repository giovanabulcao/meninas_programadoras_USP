```python
p_time1 = 0
p_time2 = 0

continuar = True

while continuar:
  srt = input()
  if srt == '1':
    p_time1 = p_time1 + 1
  elif srt == '2':
    p_time2 = p_time2 + 1


  if p_time1 >= 25 or p_time2 >= 25:
    if p_time1 > p_time2 and  p_time1 - p_time2 >= 2:
      continuar = False
    elif p_time2 > p_time1 and  p_time2 - p_time1 >= 2:
           continuar = False

print(p_time1, 'x' ,p_time2)


```

```python
parado = 1
movimento = 0
dado = input()

while dado != 'f':

   if dado == 'p':
      parado += 1
   elif dado == 'm':
      movimento += 1
   dado = input()

if movimento > parado:
  print('ativo')
else:
  print('sedentário')

```

```python
palavra = input()
letra = ''
caracteres = ''

while caracteres != '.':
  caracteres = input()
  if caracteres != '.':
    letra += caracteres

if letra == palavra:
  print('True')
else:
  print('False')



```

```python
massa = float(input())
a_inicial = float(input())
a_max = a_inicial
voltas = 0
a_atual = a_inicial

while True:
  forca = float(input())
  a = forca / massa
  voltas += 1
  if a <= a_atual:
    break
    a_atual = a

  elif a > a_max:
    a_max = a

print(f'maior acelerção: {a_max:.3f}')
print(f'número de voltas: {voltas}')
```
