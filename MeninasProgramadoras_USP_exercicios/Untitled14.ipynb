```python
vlr_disponivel = float(input())
vlr_itens = 0
saldo = vlr_disponivel

while saldo > 0:
  entrada = input()

  preco_item = float(entrada)

  if saldo >= preco_item:
    saldo = saldo - preco_item
    vlr_itens = vlr_itens + 1
  else:
    break

print(f'Número de itens {vlr_itens}')
print(f'Saldo: {saldo:.2f}')
```

```python
n_vagas = int(input())
inscritas = 0

while True:
  nome = input()
  if nome == '.':
    break
  inscritas += 1

  if inscritas < n_vagas:
    print(f'vagas sobrando: {n_vagas - inscritas}')
  elif incritas > n_vagas:
    print(f'vagas faltando: {incritas - n_vagas}')

```

```python
sinal = input()
v1 = int(input())
v2 = int(input())

if sinal == '+':
   resultado = v1 + v2
elif sinal == '-':
   resultado = v1 - v2
elif sinal == '*':
   resultado = v1 * v2
elif sinal == '**':
   resultado = v1 ** v2
elif sinal == '/':
   resultado = v1 / v2
elif sinal == '//':
   resultado = v1 // v2
elif sinal == '%':
   resultado = v1 % v2

print(resultado)

```
