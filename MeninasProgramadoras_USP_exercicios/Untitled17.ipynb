```python
vlr_alvo = int(input())
soma = 0
lista = []

while soma < vlr_alvo:
      num = input()
      for vlr in num:
        if soma >= vlr_alvo:
          break
        soma = soma + int(vlr)
        lista.append(vlr)



print(soma, lista)
```
