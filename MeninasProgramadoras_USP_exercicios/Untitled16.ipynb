```python
dpn = float(input())
lista = []

vlr = input()
while vlr != 'sair':
  lista.append(float(vlr))
  vlr = input()
comp = 0
vlr_rest = dpn
i = 0

while i < len(lista) and vlr_rest >= lista[i]:
  vlr_rest -= lista[i]
  comp += 1
  i += 1

print(f'Númeo de itens {comp}')
print(f'Saldo: {vlr_rest:.2f}')
```
